```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("downloads/mtcars.csv")
```

#1 the graph

```python
new_data = df[['wt','mpg','cyl']]
new_data.plot(kind = 'bar', color = ['g','r','b'])
```

#2 Subplots 

```python

fig, axes = plt.subplots(5)
df['mpg'].plot(ax = axes[0], kind = 'bar')
df['cyl'].plot(ax = axes[1], kind = 'bar')
df['vs'].plot(ax = axes[2], kind = 'bar')
df['disp'].plot(ax = axes[3], kind = 'bar')
df['hp'].plot(ax = axes[4], kind = 'bar')
```

#3 Scatterplots 

```python
x = df['mpg']
y = df['cyl']
plt.plot(x,y, 'o')
```

#5 Pie Chart 

```python
#settting slices 
four_cyl = df[df['cyl'] == 4]
six_cyl = df[df['cyl']== 6]
eight_cyl = df[df['cyl']== 8]

#setting portions 
four_c = len(four_cyl)
six_c = len(six_cyl)
eight_c = len(eight_cyl)
counts = np.array([four_c, six_c, eight_c])

#set labels 
cyl_labels = ["Four Cylinder", "Six Cylinder", "Eight cylinder"]
plt.pie(counts, labels = cyl_labels)
plt.show()
```
